```python
from sklearn.cluster import MeanShift

ms = MeanShift(bandwidth=60).fit(c.data["Tunnel"]["features"])
```

```python
import time
import numpy as np


start_time = time.time()

labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("cluster_centers : %d" % len(cluster_centers))
print(cluster_centers)

print("\n")

print("labels_unique : %d" % len(labels_unique))
print(labels_unique)

print("\n")

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import pandas as pd

centers = []

start_time = time.time()

for center in cluster_centers:
    result = np.where((c.data["Tunnel"]["features"] == center).all(axis=1))
    print("Result:", result[0])

    for each in result[0]:
        centers.append(each)

print("--- %s seconds ---" % (time.time() - start_time))

print(np.sort(centers))
print(len(centers))
    # for each in result[0]:
    #     centers.append(each)

```

```python
from sklearn.neighbors import KDTree

start_time = time.time()

kdt_tree = KDTree(c.data["Tunnel"]["features"])
centers = []

for i in range(len(cluster_centers)):   
    dist, ind = kdt_tree.query(cluster_centers[i].reshape(-1, 4096), k=1)
    print("Result:", dist[0], ind[0])

    index = np.ravel(ind)
    centers.append(index[0])

print(np.sort(centers))
print(len(centers))

print("--- %s seconds ---" % (time.time() - start_time))
```
